$(\cup_{i=1}^\infty A_i^\complement)^\complement$ = $\cap_i^\infty A_i^\complement$ \

Left side: \
Assume $x \in (\cup_{i=1}^\infty A_i^\complement)^\complement$. 

This means $x$ is not in the union of the complements of ($A_1, A_2, A_3, \ldots$). Therefore, $x$ is not in the complement of any of these sets. 

In other words, for all $i$, $x \notin A_i^\complement$.

This implies that $x$ is in the complement of each of the sets $A_i$, i.e., $x \in A_i^\complement$ for all $i$. 

Consequently, $x$ is in the intersection of the complements of the sets $A_1, A_2, A_3, \ldots$ which is the right side: $\cap_{i=1}^\infty A_i^\complement$.

Right side: \
Assume an element $y$ is in the RHS (intersection of complements), i.e., $y \in \cap_{i=1}^\infty A_i^\complement$. 

This means $y$ is in the complement of each of the sets $A_1, A_2, A_3, \ldots$. Therefore, $y$ is not in any of these sets, i.e., for all $i$, $y \notin A_i$.

This implies that $y$ is not in the union of these sets' originals, i.e., $y \notin \cup_{i=1}^\infty A_i$. Consequently, $y$ is in the complement of the union of the complements, which is the 
LHS: $(\cup_{i=1}^\infty A_i^\complement)^\complement$.

Since we have shown that any element from the left is also in the right and vice versa, the left side is equal to the right side. \

Hence, $(\cup_{i=1}^\infty A_i^\complement)^\complement$ = $\cap_i^\infty A_i^\complement$

*Claim:* For any two sets $A$ and $B$:
$$
|A \cup B| = |A| + |B| - |A \cap B|
$$

*Proof:* Let $|X|$ denote the cardinality of set $X$.

Consider sets $A$ and $B$.

- If $|A|$ denotes the cardinality of set $A$,

- And if $|B|$ denotes the cardinality of set $B$,

When we form the union of sets $A$ and $B$, denoted by $A \cup B$, we expect its size to be the sum of the sizes of $A$ and $B$. That is,
$$

|A \cup B| = |A| + |B|

$$

However, we need to account for the possibility that some elements might be common to both $A$ and $B$, i.e., they belong to their intersection, $A \cap B$. 

and so avoid counting these elements twice, we need to subtract the size of their intersection from the sum. If $|A \cap B|$ represents the size of the intersection set,

$$

|A \cup B| = |A| + |B| - |A \cap B|

$$

This can be expressed as a conditional statement:
If we take the union of sets $A$ and $B$, denoted by $A \cup B$, then the size of this union is equal to the sum of the sizes of $A$ and $B$, minus the size of their intersection, $A \cap B$.
